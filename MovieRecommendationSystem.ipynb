```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataset=pd.read_csv('movies.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
#selecting relevant features for recommendation
```

```python
selected_features=['genres','keywords','tagline','cast','director']
```

```python
print(selected_features)
```

```python
for feature in selected_features:
    dataset[feature]=dataset[feature].fillna('')
```

```python
#combining all the features
combined_features=dataset['genres']+' '+dataset['keywords']+' '+dataset['tagline']+' '+dataset['cast']+' '+dataset['director']
```

```python
print(combined_features)
```

```python
vectorizer=TfidfVectorizer()
combined_features=vectorizer.fit_transform(combined_features)
```

```python
print(combined_features)
```

```python
similarity=cosine_similarity(combined_features)
```

```python
print(similarity.shape)
```

```python
#getting movie name from user
movie_name=input('Enter your favourite movie name: ')
```

```python
#creating list with all movie name in dataset
list_of_all_titles=dataset['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title
index_of_the_movie=dataset[dataset.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting list of similar movies
similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
print((similarity_score))
```

```python
len(similarity_score)
```

```python
#sorting the movies based on similarityscore
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_similar_movies)
```

```python
#print the name of similar movies based on the index
print('movies suggested for you : \n')

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=dataset[dataset.index==index]['title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
        i+=1
```

```python
#movie reccomendation system
movie_name=input('Enter your favourite movie name: ')

list_of_all_titles=dataset['title'].tolist()

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)

close_match=find_close_match[0]

index_of_the_movie=dataset[dataset.title==close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print('movies suggested for you : \n')

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=dataset[dataset.index==index]['title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
        i+=1
```
